```python
df = pd.read_csv('collisions_with_coords.csv')
addresses = df.apply(lambda x : ' '.join(
    str(i).strip() for i in [x.address_number_primary, x.address_pd_primary, x.address_road_primary, x.address_sfx_primary] 
) + ', SAN DIEGO', axis = 1)
```

```python
MAPS_API_KEY = 'API KEY HERE'
coder = Geocoder(MAPS_API_KEY)

CRASHES_START = 500
CRASHES_END = 1000

locations = addresses[CRASHES_START:CRASHES_END].apply(lambda address : coder.geocode(address).coordinates)

for x in range(CRASHES_START, CRASHES_END):
    df.at[x,'lat'] = locations[x][0]
    df.at[x,'lon'] = locations[x][1]
```

```python
#Show which indecies have no lat-lon pairs yet
df['lat'].isnull()[df['lat'].isnull()]
```

```python
#DONT FORGET TO SAVE TO CSV FILE

df.to_csv('collisions_with_coords.csv')
```

```python
#converting the converted addresses to tuple and adding them to a new dataframe with 
# their respective addresses

df_locations = df
df_locations['Address']=addresses

df_locations['Coordinates'] = [(0,0) for x in range (len(addresses))]

for x in range(len(addresses)):
    df_locations.at[x, 'Coordinates'] = df.at[x, 'lat'], df.at[x, 'lon']
```

```python

df_locations = df_locations.loc[:999,['Address', 'Coordinates']]
df_locations
```
